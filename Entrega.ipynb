## **(I)** Identificación de Departamentos
* DEP_NAC: Nombre del departamento de nacimiento.
* COD_DEP_NAC: Código del departamento de nacimiento.

```python
print("Identificación de Departamentos:")
print(graduados['DEP_NAC'].unique())
print()
print(graduados['COD_DEP_NAC'].unique())
```

## **(II)** Identificación de Ciudades
* CIU_NAC: Nombre de la ciudad de nacimiento.
* COD_CIU_NAC: Código de la ciudad de nacimiento.
* LON_CIU_NAC: Longitud de la ciudad de nacimiento.
* LAT_CIU_NAC: Latitud de la ciudad de nacimiento.

```python
print("Identificación de Ciudades:")
print(graduados['CIU_NAC'].unique())
print()
print(graduados['COD_CIU_NAC'].unique())
print()
print(graduados['LON_CIU_NAC'].unique())
print()
print(graduados['LAT_CIU_NAC'].unique())
```

## **(III)** Información Geográfica General
* CODS_NAC: Código geográfico general de la región.
* CODN_NAC: Código numérico relacionado a la geografía o nacionalidad.
* NACIONALIDAD: Nacionalidad de la persona.

```python
print("Información Geográfica General:")
print(graduados['CODS_NAC'].unique())
print()
print(graduados['CODN_NAC'].unique())
print()
print(graduados['NACIONALIDAD'].unique())
```
