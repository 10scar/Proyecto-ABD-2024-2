el minimo de CD_CINE es -89 lo cual parece un error (verifiquemos los registros)

```python
df_cd_cine_neg_89 = df[df['CD_CINE'] == -89]
df_cd_cine_neg_89
```

Veamos que programas solo tienen el valor -89 o más valores

```python
unique_cd_cine_per_snies = df.groupby('SNIES_PROGRA')['CD_CINE'].unique().reset_index()
# Filtrar los registros que tienen valores negativos o más de dos valores únicos
filtered_cd_cine = unique_cd_cine_per_snies[
    unique_cd_cine_per_snies['CD_CINE'].apply(lambda x: any(val < 0 for val in x) or len(x) > 2)
]
filtered_cd_cine
```

```python
snies_progra_resultantes = filtered_cd_cine['SNIES_PROGRA'].tolist()
snies_progra_resultantes
```

Corregimos 101984 con su valor respecttvo

```python
df.loc[df['SNIES_PROGRA'] == 101984, 'CD_CINE'] = 711
```

```python
# Agrupar por SNIES_PROGRA y contar los valores únicos de CD_CINE en cada grupo
cd_cine_unique_per_snies = df.groupby('SNIES_PROGRA')['CD_CINE'].nunique().reset_index()

# Verificar si todos los SNIES_PROGRA tienen un único valor de CD_CINE
snies_determina_cd_cine = cd_cine_unique_per_snies['CD_CINE'].max() == 1

print(f"¿SNIES_PROGRA determina a CD_CINE? {snies_determina_cd_cine}")
```

```python
# Filtrar los datos para excluir los registros con CD_CINE igual a -89
df_filtered = df[df['CD_CINE'] != -89]

# Agrupar por CD_CINE y contar los valores únicos de CA_CINE en cada grupo
ca_cine_unique_per_cd_cine = df_filtered.groupby('CD_CINE')['CA_CINE'].nunique().reset_index()

# Verificar si todos los CD_CINE tienen un único valor de CA_CINE
cd_cine_determina_ca_cine = ca_cine_unique_per_cd_cine['CA_CINE'].max() == 1

print(f"¿CD_CINE determina a CA_CINE sin contar los valores con -89? {cd_cine_determina_ca_cine}")
```

```python
# Crear un conjunto de valores únicos de CD_CINE
ca_cine_values_per_program = df[df['SNIES_PROGRA'].isin(snies_progra_resultantes)].groupby('SNIES_PROGRA')['CA_CINE'].unique()
ca_cine_values_per_program
```

# Cruce de Datos Aspirantes y Matriculados 

```python
df_aspirantes = pd.read_csv('Aspirantes_y_Admitidos_mod.csv')
df_matriculados = pd.read_csv('matriculados_limpia.csv')

# Encontrar las columnas comunes
common_columns = df_matriculados.columns.intersection(df_aspirantes.columns)
common_columns_list = common_columns.tolist()

# Mostrar las columnas comunes
print("Columnas comunes:", common_columns_list)
```

Vamos a dejar solo los aspirantes que pasaron la prueba, y en el mismo rango de fecha de los matriculados 

```python
# Verificar si cada columna común tiene datos en común
for column in common_columns_list:
    common_data = df_aspirantes_admitidos[column].isin(df_matriculados[column])
    print(f"Columna: {column}")
    print(f"Datos en común: {common_data.any()}")
    print("\n")
```

Como SNIES PROGRAMA determina a programa, es más confiable usarlo, y sacar programa del calculo ya que lso datos no son en común 

```python
# Quitar la columna 'PROGRAMA' de la lista common_columns_list
common_columns_list.remove('PROGRAMA')
print(common_columns_list)
```

```python
df_aspirantes
```

```python
# Filtrar los aspirantes que fueron admitidos
df_aspirantes_admitidos = df_aspirantes[df_aspirantes['ADMITIDO'] == 'Sí']

# Obtener los rangos de YEAR y SEMESTRE de los matriculados
year_min = df_matriculados['YEAR'].min()
year_max = df_matriculados['YEAR'].max()


# Filtrar los aspirantes admitidos para que estén dentro del rango de fechas de los matriculados
df_aspirantes_admitidos = df_aspirantes_admitidos[
    (df_aspirantes_admitidos['YEAR'] >= year_min) & 
    (df_aspirantes_admitidos['YEAR'] <= year_max) 
]
df_aspirantes_admitidos
```

# Código en SQL

```python
import pandasql as psql

# Crear un ID para cada registro en df_matriculados
df_matriculados['id_matriculados'] = df_matriculados.index
df_aspirantes_admitidos['id_aspirantes'] = df_aspirantes_admitidos.index

# Realizar el merge (producto cruzado) usando una consulta SQL
query = """
SELECT 
    m.id_matriculados, 
    a.id_aspirantes, 
    m.YEAR, 
    m.SEMESTRE, 
    m.TIPO_NIVEL, 
    m.NIVEL, 
    m.DEP_NAC, 
    m.CIU_NAC, 
    m.LON_CIU_NAC, 
    m.LAT_CIU_NAC, 
    m.DEP_PROC, 
    m.CIU_PROC, 
    m.LON_CIU_PROC, 
    m.LAT_CIU_PROC, 
    m.CODS_NAC, 
    m.NACIONALIDAD, 
    m.EDAD, 
    m.SEXO, 
    m.ESTRATO, 
    m.TIPO_COL, 
    m.PBM, 
    m.MAT_PVEZ, 
    m.SEDE_NOMBRE_ADM, 
    m.SEDE_NOMBRE_MAT, 
    m.ADM_PEAMA_ANDINA, 
    m.TIPO_ADM, 
    m.PAES, 
    m.PEAMA, 
    m.MOV_PEAMA, 
    m.CONVENIO, 
    m.TIP_CONVENIO, 
    m.FACULTAD, 
    m.SNIES_PROGRA, 
    m.PROGRAMA as PROGRAMA_matriculados, 
    m.AREAC_SNIES, 
    m.CA_CINE, 
    m.CD_CINE, 
    m.AREA_CINE, 
    m.Segmento, 
    a.DEP_RES, 
    a.CIU_RES, 
    a.LON_CIU_RES, 
    a.LAT_CIU_RES, 
    a.DISCAPACIDAD, 
    a.TIPO_DISC, 
    a.TIPO_INS, 
    a.SNIES_SEDE, 
    a.INS_SEDE_NOMBRE, 
    a.PTOTAL, 
    a.ADMITIDO, 
    a.PROGRAMA as PROGRAMA_aspirantes, 
    a.MODALIDAD, 
    a.RANGO_ANO_TERMINACION
FROM 
    df_matriculados m
INNER JOIN 
    df_aspirantes_admitidos a
ON 
    """ + " AND ".join([f"m.{col} = a.{col}" for col in common_columns_list]) + """
"""

result = psql.sqldf(query, locals())

# Eliminar registros duplicados
result = result.drop_duplicates()

result.to_csv('cruce_aspirantes_matriculados.csv', index=False)
```

# Código en merge de pd

```python
# Crear un ID para cada registro en df_matriculados
df_matriculados['id_matriculados'] = df_matriculados.index
df_aspirantes_admitidos['id_aspirantes'] = df_aspirantes_admitidos.index

# Realizar el merge (producto cruzado) usando pandas
result = pd.merge(
    df_matriculados,df_aspirantes_admitidos,
    on=common_columns_list,
    how='inner', 
    suffixes=('_matriculados','_aspirantes')
)
## elimina registros duplicados 
result = result.drop_duplicates()
# Reordenar las columnas para que id_matriculados y id_aspirantes estén al principio
column_order = ['id_matriculados', 'id_aspirantes'] + [col for col in result.columns if col not in ['id_matriculados', 'id_aspirantes']]
result = result[column_order]


result.to_csv('cruce_aspirantes_matriculados.csv', index=False)
```
