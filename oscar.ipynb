# Análiss Datos Cuantitativos 

```python
quantitative_columns = df.select_dtypes(include=['int64', 'float64']).columns
quantitative_columns
```

Quitamos los datos referentes a ubocación geografica, esos van en un analisis separado

```python
columns_to_remove = ['LON_CIU_NAC', 'LAT_CIU_NAC', 'LON_CIU_PROC', 'LAT_CIU_PROC']


df_quantitative = df[quantitative_columns].drop(columns=columns_to_remove)
df_quantitative.head()
```

```python
df_quantitative.describe()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Calcular la matriz de correlación
correlation_matrix = df_quantitative.corr()

# Crear el mapa de calor
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Mapa de Calor de Correlación entre Variables Cuantitativas')
plt.show()
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 2, figsize=(12, 4))

# Ajustar los valores de los histogramas para las variables discretas
df_quantitative['YEAR'].value_counts().sort_index().plot(kind='bar', ax=axes[0], title='Distribución de Años')
df_quantitative['SEMESTRE'].value_counts().sort_index().plot(kind='bar', ax=axes[1], title='Distribución de Semestres')

plt.tight_layout()
plt.show()

# Graficar histogramas para las demás variables cuantitativas
df_quantitative.drop(columns=['YEAR', 'SEMESTRE']).hist(bins=30, figsize=(15, 10), layout=(3, 3))
plt.suptitle('Histogramas de Variables Cuantitativas')
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

```python
# Crear un boxplot para cada columna de df_quantitative
df_quantitative.plot(kind='box', subplots=True, layout=(8, 2), figsize=(20, 20), sharex=False, sharey=False)
plt.suptitle('Boxplots de Variables Cuantitativas')
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

```python
# Calcular los cuartiles y el rango intercuartílico (IQR)
Q1 = df_quantitative.quantile(0.25)
Q3 = df_quantitative.quantile(0.75)
IQR = Q3 - Q1

# Identificar los datos atípicos
outliers = ((df_quantitative < (Q1 - 1.5 * IQR)) | (df_quantitative > (Q3 + 1.5 * IQR)))

# Mostrar los datos atípicos y la cantidad de registros
for column in df_quantitative.columns:
    outlier_data = df_quantitative[column][outliers[column]]
    print(f"Columna: {column}")
    print(outlier_data.value_counts())
    print("\n")
```

```python
# Seleccionar los top 10 programas con más estudiantes
top_programas = df['SNIES_PROGRA'].value_counts().head(10).index.tolist()
df_top = df[df['SNIES_PROGRA'].isin(top_programas)]

# Boxplot del PBM por programa
plt.figure(figsize=(12, 8))
sns.boxplot(x='SNIES_PROGRA', y='PBM', data=df_top)
plt.title('Distribución del PBM por Programa Académico (Top 10)')
plt.xticks(rotation=45)
plt.xlabel('Código SNIES del Programa')
plt.ylabel('PBM')
plt.show()
```

```python
# Calcular el PBM promedio por edad
pbm_por_edad = df_quantitative.groupby('EDAD')['PBM'].mean().reset_index()

# Gráfico de líneas
sns.lineplot(x='EDAD', y='PBM', data=pbm_por_edad, marker='o', color='green')
plt.title('PBM Promedio por Edad')
plt.xlabel('Edad')
plt.ylabel('PBM Promedio')
plt.grid(True)
plt.show()
```

## Distribución Temporal (YEAR y SEMESTRE)
Objetivo: Analizar cómo varía la cantidad de estudiantes y el PBM promedio a lo largo de los años y semestres.

```python
# PBM promedio por año y semestre
pbm_data = df.groupby(['YEAR', 'SEMESTRE'])['PBM'].mean().unstack()
plt.figure(figsize=(10, 6))
sns.heatmap(pbm_data, annot=True, cmap='coolwarm')
plt.title('PBM Promedio por Año y Semestre')
plt.xlabel('Semestre')
plt.ylabel('Año')
plt.show()
```
